```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
TXT_LOCATION = r"resource/timetaken/"

def readByLine(filePath) :
    with open(filePath, 'r') as file:
        return [line.strip() for line in file]

def readAllTxtFile() :
    # list of platform name and image Name
    # platform = ["serial", "omp", "cuda"]
    platform = ["serial", "omp"]
    imName = ["lena", "wolf"]

    # dataframe 0,1,2,...,10, imageName, platform (col)
    columns = []
    
    for i in range(1, 11) :
        columns.append(i)

    columns.append("imageName")
    columns.append("platform")

    # row serial, omp, cuda 
    df = pd.DataFrame(columns=columns)

    # read data from txt file 
    i = 0
    for p in platform :
        for im in imName :
            # get txt fileName
            filePath = f"{TXT_LOCATION}{p}_{im}.txt"

            # read all the data add into dataframe 
            df.loc[i] = readByLine(filePath) + [im, p]
            i += 1
    
    return df
```

```python
import subprocess
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import json

N = 10

def readByLine(filePath) :
    with open(filePath, 'r') as file:
        return [line.strip() for line in file]

def calculateAverage(filePath) :
    result = readByLine(filePath=filePath)
    total = 0
    for i in result : 
        total += int(i)
    
    # divide by N
    return math.ceil((total / N) * 10) / 10

def resultGenerate() :
    # list of platform name and image Name
    # platform = ["serial", "omp", "cuda"]
    platform = ["serial", "omp"]
    imName = ["lena", "wolf"]
    overall = "overall"

    outBasePath = r"resource/outputForDisplay/"
    txtBasePath = r"resource/timetaken/"

    result = {}
    result["avg"] = {}

    for p in platform : 
        for im in imName : 
            # calculate overall result first
            timeOverallPath = f"{txtBasePath}{p}_{im}_{overall}.txt"
            avgOverall = calculateAverage(timeOverallPath)
            result["avg"][f"{p}_{im}_{overall}"] = avgOverall

            timeParallelPath = f"{txtBasePath}{p}_{im}.txt"
            avgParallel = calculateAverage(timeParallelPath)
            result["avg"][f"{p}_{im}"] = avgParallel
    
    # speed up
    return result
```

```python
def drawAverageRuntimeBarPlot(imName, result) : 
    x = ["serial", "omp", "cuda"]

    y = []

    for p in x : 
        y = y + result["avg"][f"{p}_{imName}"]
    
    plt.bar(x, y, color=['#7695FF', '#A2CA71', '#FF9874']) 
    plt.title(f"Average Runtime ({imName})")
    plt.xlabel("Platform")
    plt.ylabel("Average Runtime (ms)")

    for i, value in enumerate(y):
        plt.text(i, value + max(y) * 0.01, f"{value:.5f}", ha='center')

    plt.savefig(f'{imName}_avgruntime.png')
```

```python
platform = ["serial", "omp", "cuda"]

def createRuntimeDF(imageName) :
    # dataframe 0,1,2,...,10, imageName, platform (col)
    columns = []
    
    for i in range(1, 11) :
        columns.append(i)

    columns.append("imageName")
    columns.append("platform")

    # row serial, omp, cuda 
    df = pd.DataFrame(columns=columns)

    # read data from txt file 
    i = 0
    for p in platform :
        # get txt fileName
        filePath = f"{TXT_LOCATION}{p}_{imageName}.txt"

        # read all the data add into dataframe 
        df.loc[i] = readByLine(filePath) + [imageName, p]
        i += 1
    
    return df

def drawRuntimeLineGraph(df, imageName, basePath) :
    # melt the dataframe so columns 1 - 10 become a single 'X' col, time usage go into 'Y'
    df_melted = df.melt(
        id_vars=['imageName', 'platform'],
        var_name='X',          # was "1","2","3" etc.
        value_name='Y'         # time usage
    )

    df_melted['X'] = pd.to_numeric(df_melted['X'])

    plt.figure(figsize=(9,6))
    sns.lineplot(
        data=df_melted,
        x='X',
        y='Y',
        hue='platform',
        marker='o'
    )
    # Optionally label points
    for i in range(len(df_melted)):
        x_val = df_melted.loc[i, 'X']
        y_val = df_melted.loc[i, 'Y']
        plt.text(x_val+0.1, y_val, str(y_val), fontsize=8)
    
    plt.title(f"Serial vs. OMP vs. CUDA Runtimes ({imageName})")
    plt.xlabel("Column (1 to 10)")
    plt.ylabel("Time Used")
    plt.legend(title="Platform")
    plt.show()
    plt.savefig(f'{basePath}{imageName}_runtime.png')
```
